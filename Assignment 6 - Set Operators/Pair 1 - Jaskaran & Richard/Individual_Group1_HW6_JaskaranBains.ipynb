**Proposition 1**

"Find the out of stock items"

This query takes the product ids from orders and excludes the ones from the productids in inventory to show the out of stock items.

\- Firstly we use a subquery with the set operator EXCEPT to get the productids that are in every sale, and exclude those that are found in the inventory. This way we return a column of product ids which are sold out

\- We assign the alias "OutOfStockProducts" to this subquery. 

\- In the main query we use a JOIN clause to take the subquery and connect it to the products table on productID of the products table to productID of the subquery. 

\- THe purpose of the JOIN is since our subquery only returns product IDs we also would like the name of the product so our JOIN allows us to see the name of the products.

```sql
SELECT p.ProductID, p.Name AS ProductName
FROM Production.Product p
JOIN (
    SELECT ProductID
    FROM Sales.SalesOrderDetail
    EXCEPT
    SELECT ProductID
    FROM Production.ProductInventory
) AS OutOfStockProducts
ON p.ProductID = OutOfStockProducts.ProductID;
```

**Proposition 2**

"Find the usernames of our customers that have emails with us"

This query checks if a person has an email address in the db and then removes the website part of it and returns the "Username"

\- Firstly we use a subquery, within the subqery is the UNION operator for the BusinessEntityID from the table of emails and personID from the table of customers.

\- Within the main query we use SELECT to define our output of the query as the business entity ID which we alias as PersonID because its the same thing but one is easier read and more intuitive. 

\- In SELECT we also utilize REPLACE(column, substring, replacementsubstring) which is a substring function that allows us to replace instances of a specified substring with a different one.  In this case we are removing the suffix of the email by replacing it with a space. We alias the output of the replace as Username. 

\- We use the WHERE to find the PersonID in the union of the tables within the subquery to acquire those that Exist, but with an Email in our db.

```sql
SELECT 
    ea.BusinessEntityID AS PersonID,
    REPLACE(ea.EmailAddress, '@adventure-works.com', '') AS Username
FROM Person.EmailAddress ea
WHERE ea.BusinessEntityID IN (
    SELECT BusinessEntityID
    FROM Person.EmailAddress
    UNION
    SELECT PersonID
    FROM Sales.Customer
);
```

**Proposition 10**

"Find products that dont exist in any other table"  

\- First we use a sub-subquery, within the sub-subquery is the UNION of all products that exist in the salesorderdetail table, the purchaseorderdetail table, and the bill of materials table.   
  
\- The outer subquery seelcts the the Product table that has the list of all products. This outer subquery then takes this table and with the EXCEPT operator, it removes those items that are in the sub sub query from the items of its Products table. Our subquery is aliased as inactiveProducts

<span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">- In our main query, we use a JOIN to connect the results of our sub and sub query to our Products table ON their products IDs so that when we output them we can reveal the productId along with the name of the product.&nbsp;</span> 

\- Our singular output is the only product that does not appear on any other table so is completely inactive.

```sql
SELECT p.ProductID, p.Name
FROM Production.Product p
JOIN(
    SELECT ProductID
    FROM Production.Product
    EXCEPT
    (
    SELECT ProductID
    FROM Sales.SalesOrderDetail
    UNION
    SELECT ProductID
    FROM Purchasing.PurchaseOrderDetail
    UNION
    SELECT ComponentID
    FROM Production.BillOfMaterials
    )
    ) AS inactiveProducts 
ON inactiveProducts.ProductID = p.ProductID;
```
