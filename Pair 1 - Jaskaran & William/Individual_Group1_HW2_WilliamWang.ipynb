<span style="font-size: 13pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-weight: 700; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">PROPOSITION 1&nbsp;</span> 

  

<span style="font-size: 11pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-style: italic; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">"Make a list of the 10 most populated countries."&nbsp;</span> 

  

<span style="font-size: 11pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">The query examines all countries, finds their populations, and selects the top 10 with the highest numbers. It then arranges them in descending order (largest to smallest) using the DESC keyword so we can identify the countries with the highest populations.</span>

<span style="font-size: 11pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;"><br></span>

```sql
SELECT TOP (10) CountryID, CountryName, LatestRecordedPopulation
FROM WideWorldImporters.Application.Countries
ORDER BY LatestRecordedPopulation DESC
```

### <span style="font-size: 13pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-weight: 700; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">PROPOSITION 2</span>

<span style="font-size: 11pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-style: italic; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">"Find those who share the same name, record the count, and do a ranking from most common to least common repeated name."</span>

<span style="font-size: 14.6667px; white-space-collapse: preserve;">The query counts the occurrences of PreferredName in the People table. We use GROUP BY to count names and HAVING COUNT(*) &gt; 1 to exclude unique names. Sorting in descending order ranks names from most to least common.</span>  

<span style="font-size: 11pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;"><br></span>

```sql
SELECT PreferredName, COUNT(*) AS NameCount
FROM WideWorldImporters.Application.People 
GROUP BY PreferredName
HAVING COUNT(*) > 1
ORDER BY NameCount DESC;
```

### <span style="font-size: 13pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-weight: 700; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">PROPOSITION 3</span>

<span style="font-size: 11pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-style: italic; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">"Order the suppliers based on the number of orders they have, where the supplier with the most orders is ordered first."</span>

<span style="font-size: 11pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">This query counts the number of orders per supplier from the PurchaseOrders table. We use GROUP BY to group orders per supplier and count them. Ordering by OrderCount DESC ensures suppliers with the most orders appear first.</span>

<span style="font-size: 11pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;"><br></span>

```sql
SELECT SupplierID, COUNT(PurchaseOrderID) AS OrderCount
FROM WideWorldImporters.Purchasing.PurchaseOrders
GROUP BY SupplierID
ORDER BY OrderCount desc;
```

<span style="font-size: 11pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-style: italic; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">"Make a list of the 10 most populated cities in the United States."</span>

<span style="font-size: 11pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-style: italic; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">"Make a list of people who have at least 4 ‘e’s in their name."</span>

<span style="font-size: 14.6667px; white-space-collapse: preserve;">The LIKE clause searches for names containing at least four ‘e’s. Filtering is done using wildcard matching.</span>

### **PROPOSITION 6**

_"Find those who placed their orders at the end of the month. From most recent to least recent."_

**EOMONTH(OrderDate) filters for end-of-month orders. Ordering by OrderDate DESC ensures the latest orders appear first.**

```sql
SELECT OrderID,OrderDate,CustomerID
FROM WideWorldImporters.Sales.Orders
WHERE OrderDate = EOMONTH(OrderDate)
ORDER BY OrderDate desc
```

### <span style="font-size: 13pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-weight: 700; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">PROPOSITION 7</span>

<span style="font-size: 11pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-style: italic; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">"Find the person who placed the most orders in December 2015."</span>

<span style="font-size: 14.6667px; white-space-collapse: preserve;">This query filters orders between Dec 1 - Dec 31, 2015.</span>

<span style="font-size: 14.6667px; white-space-collapse: preserve;">GROUP BY ContactPersonID counts the orders per person.</span>

<span style="font-size: 14.6667px; white-space-collapse: preserve;">Ordering by BuyCount DESC ensures the person with the most orders appears first.</span>

```sql
SELECT TOP(1) ContactPersonID, COUNT(*) AS BuyCount
FROM WideWorldImporters.Sales.Orders
WHERE orderdate BETWEEN '2015-12-01' AND '2015-12-31'
GROUP BY ContactPersonID
HAVING COUNT(*) > 0
ORDER BY BuyCount desc
```

### <span style="font-size: 13pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-weight: 700; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">PROPOSITION 8</span>

<span style="font-size: 11pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-style: italic; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">"Find orders picked up by the carrier between January and February 2013."</span>

<span style="font-size: 14.6667px; white-space-collapse: preserve;">The WHERE clause filters orders completed between January and February 2013.</span>

<span style="font-size: 14.6667px; white-space-collapse: preserve;">Ordering ensures data is structured by completion time.</span>

```sql
SELECT PickedByPersonID, PickingCompletedWhen
FROM WideWorldImporters.Sales.Orders
WHERE PickingCompletedWhen BETWEEN '2013-01-01 12:00:00.0000000' AND '2013-02-28 12:00:00.0000000'
```

<span style="font-size: 11pt; font-family: Arial, sans-serif; color: rgb(0, 0, 0); background-color: transparent; font-style: italic; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">"Find items that are for children that are not over 20kg."</span>

### **PROPOSITION 10**

_"Find the state with the longest name."_

The LEN function measures name length. 

Sorting by DESC ensures the longest state name appears first.

```sql
SELECT TOP(1) StateProvinceName, LEN(StateProvinceName) AS NameLength
FROM WideWorldImporters.Application.StateProvinces
ORDER BY NameLength DESC

```
