```sql
---- Exercise 6 ------
-- 6
-- Explain the difference between the following two queries

-- Query 1
USE TSQLV4
SELECT empid, COUNT(*) AS numorders
FROM Sales.Orders
WHERE orderdate < '20160501'
GROUP BY empid;

-- Query 2
USE TSQLV4
SELECT empid, COUNT(*) AS numorders
FROM Sales.Orders
GROUP BY empid
HAVING MAX(orderdate) < '20160501';
```

Although these 2 queries seem like they're proposing the same thing but since we get different results , (4 rows instead of 9, where empid 1,2,4,7, and 8 are excluded) our next question is why ? A closer look reveals that in query 2 our initially queried data is filtered out by the HAVING clause with the attribute **MAX** of a date meaning that in query 1 we're presented with empids that have any order before the given date 20160501 and in query 2 we are presented only those empids that have _only_ order before the date 20160501. So our "initial" collection is the same but gets refiltered in query 2.

```sql
---- Exercise 7 ------
-- 7 
-- Return the three ship countries with the highest average freight for orders placed in 2015
-- Tables involved: Sales.Orders table

-- Desired output:

shipcountry     avgfreight
--------------- ---------------------
Austria         178.3642
Switzerland     117.1775
Sweden          105.16

(3 row(s) affected)
```

```sql
USE TSQLV4
SELECT TOP 3 shipcountry, AVG(freight) AS avgfreight
FROM Sales.Orders
WHERE orderdate >= '20150101' AND orderdate < '20160101'
GROUP BY shipcountry
ORDER BY avgfreight DESC;
```

**PROPOSITION 1**

"Show me the top 100 least populated cities ordered ascendingly from the smallest population down to the highest within that top 100"

The query itself is simple enough, TOP (100), used to acquire the top 100 in ranking ascending is applied by default.

We are using an alias for LatestRecordedPopulation for no other reason that its long and hard to read so "Population" suffices.

We use FROM to grab our specific schema.

<span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">We use WHERE clause "&gt;1" to exclude any NULL rows.</span>

<span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">We use ORDER BY just standardly to sort by population ascendingly.</span>

```sql
SELECT TOP (100) CityID,
                 StateProvinceID,
                 CityName, 
                 LatestRecordedPopulation AS Population
FROM [WideWorldImporters].[Application].[Cities]
WHERE LatestRecordedPopulation > 1
ORDER BY Population
```

**PROPOSITION 2**

"Show me all the orders from individual customers in the year 2016 and sort them by which customer has the most orders descendingly"

This query uses only 2 columns, just CustomerID and OrderDate where OrderDate is counted, and is aliased to "NumberofOrders".

We use FROM to grab our specific schema.

<span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">We use WHERE clause to specify the specific year we are concerned about.</span>

<span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">We use ORDER BY to sort based on the amount of orders.</span>

```sql
SELECT CustomerID, 
       COUNT(OrderDate) AS NumberOfOrders
FROM [WideWorldImporters].[Sales].[Orders]
WHERE YEAR(OrderDate) = 2016
GROUP BY CustomerID
ORDER BY NumberOfOrders DESC;
```

**PROPOSITION 9**

"Show the delivery methods used and how many suppliers use those methods"

In this query we are trying to get a table that shows Delivery Methods next to the amount of suppliers that use it  
first we implement the tables we need which in this case is DeliveryMethod and we are selecting a count function star to show  
we are counting members of the data base _Suppliers_ and putting those numbers into our table NumberOfSuppliers.

We are using the clause WHERE only to remove NULLS because we do not know what the NULL means. 

We are using GROUP BY clause to essentially make groups of each delivery method and then everytime the count sees that delivery   
method being used we increment the NumberofSuppliers for that row and do this for each method.

We are ordering by NumberOfSuppliers descendingly so that we can quickly see the most used DeliveryMethod.

```sql
SELECT DeliveryMethod, COUNT(*) AS NumberOfSuppliers
FROM WideWorldImporters.Website.Suppliers
WHERE DeliveryMethod IS NOT NULL
GROUP BY DeliveryMethod 
ORDER BY NumberOfSuppliers DESC;
```
