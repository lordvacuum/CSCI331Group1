```
The query identifies the  most ordered stock item  by calculating the total quantity ordered for each product in the Sales.OrderLines table and then selecting the  top-selling  item from Warehouse.StockItems. It ensures that only the item with the highest total quantity ordered is returned.
```

````sql
USE WideWorldImporters;

WITH ProductOrderCounts AS (
    SELECT 
        o.StockItemID, 
        SUM(o.Quantity) AS TotalQuantity
    FROM Sales.OrderLines AS o
    GROUP BY o.StockItemID
)
SELECT TOP 1 
    s.StockItemID, 
    s.StockItemName, 
    p.TotalQuantity
FROM Warehouse.StockItems AS s
INNER JOIN ProductOrderCounts AS p
ON s.StockItemID = p.StockItemID
ORDER BY p.TotalQuantity DESC;
````

<span style="font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">The query retrieves </span> <span style="font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); background-color: transparent; font-weight: 700; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">customers who have never placed an order</span> <span style="font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;"> by using a </span> <span style="font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); background-color: transparent; font-weight: 700; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">LEFT JOIN</span> <span style="font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;"> between Sales.Customers and a subquery that selects distinct CustomerIDs from Sales.Orders. The WHERE ActiveCustomers.CustomerID IS NULL condition filters out customers who have placed at least one order, leaving only those without any orders.</span>

````sql
USE WideWorldImporters;

SELECT C.CustomerID, C.CustomerName
FROM Sales.Customers AS C
LEFT JOIN (
    SELECT DISTINCT CustomerID FROM Sales.Orders
) AS ActiveCustomers
ON C.CustomerID = ActiveCustomers.CustomerID
WHERE ActiveCustomers.CustomerID IS NULL;
````

<span style="font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">The query retrieves each customers CustomerID along with the </span> <span style="font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); background-color: transparent; font-weight: 700; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">total number of orders they have placed</span> <span style="font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;"> by performing a </span> <span style="font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); background-color: transparent; font-weight: 700; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">LEFT JOIN</span> <span style="font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;"> between Sales.Customers and Sales.Orders. Customers who have not placed any orders will still be included in the result, with numorders displayed as </span> <span style="font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); background-color: transparent; font-weight: 700; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;">0</span> <span style="font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); background-color: transparent; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline; white-space-collapse: preserve;"> due to the COUNT(O.OrderID).</span>

````sql
USE WideWorldImporters;

SELECT C.CustomerID, COUNT(O.orderid) AS numorders
FROM Sales.Customers AS C
LEFT OUTER JOIN Sales.Orders AS O
ON C.CustomerID = O.CustomerID
GROUP BY C.CustomerID;
````

<span style="background-color: transparent; white-space-collapse: preserve; font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline;">The query retrieves orders where the OrderDate is </span> <span style="background-color: transparent; white-space-collapse: preserve; font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); font-weight: 700; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline;">not the last day of the year</span> <span style="background-color: transparent; white-space-collapse: preserve; font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline;"> by calculating the </span> <span style="background-color: transparent; white-space-collapse: preserve; font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); font-weight: 700; font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline;">end-of-year date</span> <span style="background-color: transparent; white-space-collapse: preserve; font-size: 12pt; font-family: &quot;Times New Roman&quot;, serif; color: rgb(0, 0, 0); font-variant-numeric: normal; font-variant-east-asian: normal; font-variant-alternates: normal; font-variant-position: normal; vertical-align: baseline;"> (December 31st) for each order using DATEFROMPARTS. It then filters out orders that fall on this date, ensuring only orders made on other days are included.</span>.

````sql
USE WideWorldImporters;

WITH C AS (
  SELECT *,
    DATEFROMPARTS(YEAR(OrderDate), 12, 31) AS EndOfYear
  FROM Sales.Orders
)
SELECT OrderDate, CustomerID, EndOfYear
FROM C
WHERE OrderDate <> EndOfYear;
````
