Proposition 4 "Query that returns all orders placed by the customers who placed the highest number of order"

````sql
--code 4
USE WideWorldImporters
GO
SELECT CustomerID, OrderID
FROM Sales.Orders
WHERE CustomerID IN (
    SELECT CustomerID
    FROM Sales.Orders
    GROUP BY CustomerID
    HAVING COUNT(OrderID) = (
        SELECT MAX(MaxOrderCount)
        FROM (
            SELECT CustomerID, COUNT(OrderID) AS MaxOrderCount
            FROM Sales.Orders
            GROUP BY CustomerID
        ) AS OrderCounts
    )
);
````

Propistion 6 "sales per year"

````sql
--code 6 -

USE WideWorldImporters
GO 
WITH YearlySales AS   (
    SELECT 
        YEAR(od.OrderDate) AS SalesYear, --column alias
        COUNT(od.OrderID) AS TotalSales,   --CA
        SUM(ol.Quantity) AS TotalQuantity  --CA
    FROM Sales.Orders od
    JOIN Sales.OrderLines ol ON od.OrderID = ol.OrderID
    GROUP BY YEAR(od.OrderDate)
)
SELECT * FROM YearlySales
ORDER BY SalesYear DESC;
````

Proposition 9 "<span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">Create a view that returns the total quantity for each customer and year</span>"

````sql
--code 9

USE WideWorldImporters
DROP VIEW IF EXISTS CustomerYearlySales;
GO

CREATE VIEW CustomerYearlySales AS
WITH YearlySales AS (
    SELECT 
        od.CustomerID,
        YEAR(od.OrderDate) AS SalesYear,
        SUM(ol.Quantity) AS TotalQuantity
    FROM Sales.Orders od
    JOIN Sales.OrderLines ol ON od.OrderID = ol.OrderID
    GROUP BY od.CustomerID, YEAR(od.OrderDate)
)
SELECT * FROM YearlySales;
````

Proposition 10 "Return each customer's most recent order details and expecteddeliverydates"Â **\--cross apply pg 198**

````sql
--code 10
USE WideWorldImporters
GO  
SELECT 
    c.CustomerID, 
    c.CustomerName, 
    o.OrderID, 
    o.OrderDate, 
    o.ExpectedDeliveryDate
FROM Sales.Customers c
CROSS APPLY (
    SELECT TOP 1 
        o.OrderID, 
        o.OrderDate, 
        o.ExpectedDeliveryDate
    FROM Sales.Orders o
    WHERE o.CustomerID = c.CustomerID
    ORDER BY o.OrderDate DESC
) o;
````
